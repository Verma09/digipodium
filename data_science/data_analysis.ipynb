- Selecting rows and cols
  - rows(slicing)
      - 'loc' - gets us the data using named indexes
      - 'iloc' - gets us the data using numeric indexes
   - cols (dictionary style)
       - 'df[single]'- requires name of column
       - 'df[multiple]'- requirs a list columns(multiple is list of named cols)

```python
# single row (row)
weather.iloc[0]
```

```python
weather.iloc[-1]
```
